```python
from s2sphere import CellId, RegionCoverer, LatLng, LatLngRect, Cell
import random
# 创建一个 RegionCoverer 实例
coverer = RegionCoverer()

# 设置要划分的级别
level = 0 # 选择您希望的级别
coverer.min_level = level
coverer.max_level = level

# 定义一个经纬度范围，这里表示全球
rect = LatLngRect.full()

# 使用 RegionCoverer 获取 S2 单元列表
covering = coverer.get_covering(rect)

# 创建一个集合来记录出现过的 face 值
unique_faces = set()
print(len(covering))
# 遍历 S2 单元列表并打印信息，同时记录出现过的 face 值
for cell_id in covering:
    cell = CellId(cell_id.id())
    face = cell.face()
    unique_faces.add(face)
    print(f"Cell ID: {cell.id()}")
    print(f"Level: {cell.level()}")
    print(f"Face: {face}")
    print(f"Center Coordinates: {cell.to_lat_lng()}")
    print("---")

# 打印出现过的 face 值
print("Unique Faces:")
for face in unique_faces:
    print(f"Face {face}")
```

```python
import csv
from s2sphere import CellId, RegionCoverer, LatLng, LatLngRect, Cell

# 创建一个 RegionCoverer 实例
coverer = RegionCoverer()

# 设置要划分的级别
level = 4  # 选择您希望的级别
coverer.min_level = level
coverer.max_level = level

# 定义一个经纬度范围，这里表示全球
rect = LatLngRect.full()

# 使用 RegionCoverer 获取 S2 单元列表
covering = coverer.get_covering(rect)

# 创建一个集合来记录出现过的 face 值
unique_faces = set()
print(len(covering))

# 准备将数据写入CSV文件
output_file = f"S2_cells_level_{level}.csv"

with open(output_file, mode='w', newline='') as file:
    csv_writer = csv.writer(file)
    
    # 写入CSV文件的标题行，按指定的顺序
    csv_writer.writerow(["Class_Label", "Hex_ID", "Latitude_Mean", "Longitude_Mean"])

    # 初始化 class_label 为 0
    class_label = 0
    
    # 遍历S2单元列表并写入CSV文件
    for cell_id in covering:
        cell = CellId(cell_id.id())
        face = cell.face
        
        # 计算中心坐标
        center = cell.to_lat_lng()
        latitude_mean = center.lat().degrees
        longitude_mean = center.lng().degrees
        
        # 获取 Hex_ID
        hex_id = cell.to_token()
        
        # 将数据写入CSV文件，按指定的顺序
        csv_writer.writerow([class_label, hex_id, latitude_mean, longitude_mean])
        
        # 增加 class_label
        class_label += 1
```

```python
import os
from s2sphere import CellId, LatLng, LatLngRect, RegionCoverer,Cell
import simplekml

# 创建一个 RegionCoverer 实例
coverer = RegionCoverer()

# 设置要划分的级别
level = 0  # 选择您希望的级别
coverer.min_level = level
coverer.max_level = level

# 定义一个经纬度范围，这里表示全球
rect = LatLngRect.full()

# 使用 RegionCoverer 获取 S2 单元列表
covering = coverer.get_covering(rect)

# 创建一个 KML 对象
kml = simplekml.Kml()

# 遍历S2单元列表并添加每个S2单元的边界到KML文件
for cell_id in covering:
    cell = CellId(cell_id.id())
    vertices = [Cell(cell).get_vertex(v) for v in range(4)]
    
    list1 = [LatLng.from_point(vertices[i]) for i in range(4)]
    print(list1)
    coords = [(v.lat().degrees, v.lng().degrees) for v in list1]
    # 创建一个多边形
    pol = kml.newpolygon(name=cell.to_token())
    pol.outerboundaryis = coords
    pol.style.polystyle.color = simplekml.Color.blue

# 保存KML文件
kml.save("s2_cells.kml")
```
